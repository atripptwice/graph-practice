### Basic Graph

```python
#Make x and y 
x = [i for i in range(0,10)]
y = [i for i in range(0,20,2)]
x_tick_list = [i for i in range(0,11,)]
y_tick_list = [i for i in range(0,22,2)]
#formatting 
title_font = {'fontsize':20,'fontname':'Arial','color':'Purple'}
axis_font = {'fontsize':13,'fontname':'Arial','color':'Purple'}
#using the short hand fmt [marker][line][color] exlucding marker we have too many points so it looks weird
plt.plot(x,y,'x--c',label="Thing being plotted")
#line number 2
x2 = np.arange(0,5,.5)

plt.plot(x2,x2**2,'o-r',label='X Squared')

plt.xticks(x_tick_list)
plt.yticks(y_tick_list)
plt.tick_params(color="Green",length= 10,width =2)
plt.title("My First Graph (not really..)",fontdict= title_font)
plt.xlabel("X Axis",fontdict = axis_font)
plt.ylabel("Y Axis",fontdict= axis_font)
plt.legend()
# show plot
plt.show()
```
